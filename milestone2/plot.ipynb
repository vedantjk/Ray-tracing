```python
import numpy as np
import matplotlib.pyplot as plt

# Read data from "output.txt"
data = np.loadtxt('output.txt')

# Create a 2D grid
X, Y = np.meshgrid(np.linspace(-10, 10, 512), np.linspace(-10, 10, 512))

# Plot the intensity values
plt.imshow(data, extent=(-10, 10, -10, 10), origin='lower', cmap='gray')
plt.colorbar(label='Intensity')
plt.title('Sphere Rendering')
plt.xlabel('X')
plt.ylabel('Z')
plt.show()
```
